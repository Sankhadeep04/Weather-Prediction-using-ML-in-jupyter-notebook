```python
def classify_weather(row):
    if row['temp_max'] <= 2.5 and row['precipitation'] > 0.2:
        return "Snowy"
    elif row['precipitation'] > 0.5 and row['temp_max'] > 2.5:
        return "Rainy"
    elif row['sunshine'] >= 6 and row['cloud_cover'] <= 3 and row['humidity'] < 0.8:
        return "Sunny"
    elif row['cloud_cover'] >= 6 and row['precipitation'] <= 0.5:
        return "Cloudy"
    else:
        return "Cloudy"

# Apply function to dataset
weather['weather'] = weather.apply(classify_weather, axis=1)

# Check first 10 rows with all important fields including MONTH
print(weather.head(10)[['DATE','MONTH','temp_min','temp_max','precipitation',
                        'sunshine','cloud_cover','humidity','weather']])
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score

# ------------------------------
# Step 1: Features and Target
# ------------------------------
X = weather[['MONTH','cloud_cover','humidity','precipitation',
             'sunshine','temp_mean','temp_min','temp_max']]
y = weather['weather_type']

# Encode labels
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# ------------------------------
# Step 2: Train-Test Split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# ------------------------------
# Step 3: Apply SMOTE on Training Data Only
# ------------------------------
smote = SMOTE(random_state=42, k_neighbors=2)  # k_neighbors=2 because Snowy has very few samples
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", dict(zip(*np.unique(y_train, return_counts=True))))
print("After SMOTE :", dict(zip(*np.unique(y_train_res, return_counts=True))))

# ------------------------------
# Step 4: Train Random Forest
# ------------------------------
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_res, y_train_res)

# ------------------------------
# Step 5: Evaluate Model
# ------------------------------
y_pred = rf.predict(X_test)

print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred, target_names=encoder.classes_))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Features & target
X = weather[['MONTH','cloud_cover','humidity','precipitation',
             'sunshine','temp_mean','temp_min','temp_max']]
y = weather['weather_type']

# Encode target labels
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)

# Stratified K-Fold setup
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold = 1
acc_scores = []
all_reports = []

for train_idx, test_idx in skf.split(X, y_encoded):
    # Split data
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y_encoded[train_idx], y_encoded[test_idx]
    
    # Apply SMOTE only on training data
    smote = SMOTE(random_state=42)
    X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
    
    # Train Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X_train_res, y_train_res)
    
    # Predictions
    y_pred = rf.predict(X_test)
    
    # Accuracy & Report
    acc = accuracy_score(y_test, y_pred)
    acc_scores.append(acc)
    report = classification_report(y_test, y_pred, target_names=encoder.classes_, output_dict=True)
    all_reports.append(report)
    
    print(f"\n📂 Fold {fold} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred, target_names=encoder.classes_))
    
    fold += 1

# Average Accuracy
print("\n✅ Mean Accuracy across folds:", np.mean(acc_scores))
```

```python
import pandas as pd
import numpy as np

# --- 0) Make sure existing types are right and fix existing temp_mean NaNs ---
weather['date'] = pd.to_datetime(weather['date'], errors='coerce')
weather = weather.sort_values('date').reset_index(drop=True)

# If any existing temp_mean is NaN, compute it from min/max
mask_mean_nan = weather['temp_mean'].isna()
weather.loc[mask_mean_nan, 'temp_mean'] = (
    (weather.loc[mask_mean_nan, 'temp_min'] + weather.loc[mask_mean_nan, 'temp_max']) / 2
)

# --- 1) Build 20 NEW Snowy rows that satisfy your rule ---
# Rule requires: temp_max <= 2.5 and precipitation > 0.2
start_date = (weather['date'].max() + pd.Timedelta(days=1)) if not weather['date'].isna().all() else pd.Timestamp('2017-01-01')
new_dates = pd.date_range(start=start_date, periods=20, freq='D')

rng = np.random.default_rng(42)
snowy_rows = pd.DataFrame({
    'date': new_dates,
    'MONTH': new_dates.month,                            # keep MONTH in sync with date
    'cloud_cover': rng.integers(6, 9, size=20),         # 6..8 (cloudy scale 0..8)
    'humidity': rng.uniform(0.75, 0.95, size=20),       # high humidity (0..1 scale)
    'precipitation': rng.uniform(0.30, 2.00, size=20),  # > 0.2 to trigger Snowy
    'sunshine': rng.uniform(0.0, 3.0, size=20),         # low sunshine during snow
    'temp_min': rng.uniform(-10.0, -2.0, size=20),      # cold
    'temp_max': rng.uniform(-5.0,  2.5, size=20),       # must be <= 2.5
})

# Enforce the Snowy rule strictly
snowy_rows['temp_max'] = snowy_rows['temp_max'].clip(upper=2.5)
snowy_rows['precipitation'] = snowy_rows['precipitation'].clip(lower=0.21)

# Compute temp_mean for the new rows
snowy_rows['temp_mean'] = (snowy_rows['temp_min'] + snowy_rows['temp_max']) / 2

# Set the label
snowy_rows['weather_type'] = 'Snowy'

# --- 2) Append to your dataset ---
weather = pd.concat([weather, snowy_rows], ignore_index=True)

# --- 3) Round ALL float columns to 2 decimals (max two digits after dot) ---
float_cols = weather.select_dtypes(include=['float64', 'float32']).columns
weather[float_cols] = weather[float_cols].round(2)

# --- 4) Store 'date' without time part (no 00:00:00) ---
# Convert entire column to date objects so display has no time component
weather['date'] = pd.to_datetime(weather['date']).dt.date

# --- 5) Final sanity checks ---
# Recompute any remaining temp_mean NaNs (should be none)
if weather['temp_mean'].isna().any():
    mask = weather['temp_mean'].isna()
    weather.loc[mask, 'temp_mean'] = ((weather.loc[mask, 'temp_min'] + weather.loc[mask, 'temp_max']) / 2).round(2)

# Optional: check the last few rows
print(weather.tail(25)[['date','MONTH','temp_min','temp_max','temp_mean','precipitation','sunshine','cloud_cover','humidity','weather_type']])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

# -----------------------------
# 1. Separate features and target
# -----------------------------
X = weather.drop(columns=['weather_type', 'date'])   # features
y = weather['weather_type']                          # target

# -----------------------------
# 2. Define Stratified K-Fold
# -----------------------------
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Model to use
model = RandomForestClassifier(random_state=42)

# Store fold results
fold_accuracies = []

# -----------------------------
# 3. Run Cross-Validation
# -----------------------------
for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # --- Optional SMOTE balancing ---
    sm = SMOTE(random_state=42)
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    
    # --- Train ---
    model.fit(X_train_res, y_train_res)
    
    # --- Predict ---
    y_pred = model.predict(X_test)
    
    # --- Evaluate ---
    acc = accuracy_score(y_test, y_pred)
    fold_accuracies.append(acc)
    
    print(f"Fold {fold} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred, zero_division=0))

# -----------------------------
# 4. Final Result
# -----------------------------
# print(f"\nMean CV Accuracy: {np.mean(fold_accuracies):.4f}")
mean_accuracy = np.mean(fold_accuracies)
print(f"\nMean CV Accuracy: {mean_accuracy:.4f} ({mean_accuracy*100:.2f}%)")
```

```python
def predict_weather_cases(model, cases):
    results = []
    for case in cases:
        temp_mean = round((case["temp_min"] + case["temp_max"]) / 2, 2)

        # Create dataframe in correct feature order
        input_data = pd.DataFrame([{
            'MONTH': case["MONTH"],
            'cloud_cover': case["cloud_cover"],
            'humidity': case["humidity"],
            'precipitation': case["precipitation"],
            'sunshine': case["sunshine"],
            'temp_mean': temp_mean,
            'temp_min': case["temp_min"],
            'temp_max': case["temp_max"]
        }])[model.feature_names_in_]

        prediction = model.predict(input_data)[0]
        results.append((case, prediction))
    return results
```
